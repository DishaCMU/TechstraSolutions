```python
df1 = pd.read_csv('company_reviews_glassdoor_ADP.csv',encoding='latin')
df2 = pd.read_csv('company_reviews_glassdoor_Amazon.csv',encoding='latin')
df3 = pd.read_csv('company_reviews_glassdoor_American-express.csv',encoding='latin')
df4 = pd.read_csv('company_reviews_glassdoor_earnst-and-young.csv',encoding='latin')
df5 = pd.read_csv('company_reviews_glassdoor_Google.csv',encoding='latin')
df6 = pd.read_csv('company_reviews_glassdoor_Highmark-Health.csv',encoding='latin')
df7 = pd.read_csv('company_reviews_glassdoor_Microsoft.csv',encoding='latin')
df =pd.concat([df1, df2,df3, df4,df5, df6,df7])
df.drop('Unnamed: 0', axis = 1, inplace = True)
df
```

```python
posReviews =df['posReview']
negReviews = df['negReviews']
```

```python

# library to clean data 
import re  
  
# Natural Language Tool Kit 
import nltk  
  
nltk.download('stopwords') 
  
# to remove stopword 
from nltk.corpus import stopwords 
  
import string
```

```python

 
lemmatizer=nltk.stem.wordnet.WordNetLemmatizer()
# 1000 (reviews) rows to clean 

def makeCorpus(reviews):
    corpus = [] 
    for review in reviews:  

        # column : "Review", row ith 
        #review = re.sub('[^a-zA-Z]', ' ', posReview)  

        # convert all cases to lower cases 
        review = review.lower()  

        # split to array(default delimiter is " ") 
        review = review.split()  



        # loop for stemming each word 
        # in string array at ith row     
        review = [lemmatizer.lemmatize(word) for word in review 
                    if not word in set(stopwords.words('english'))]  

        # rejoin all string array elements 
        # to create back into a string
        review1=[]
        for word in review:
            for ch in string.punctuation:                                                                                                     
                word = word.replace(ch, "")

            review1.append(word)
        review1 = [x for x in review1 if x]
        #print(review1)


        # append each string to create 
        # array of clean text  
        corpus.append(review1)  
    return corpus
```

```python
posCorpus = makeCorpus(posReviews)
negCorpus = makeCorpus(negReviews)


```

```python
##downloading the nltk packages
import nltk
nltk.download('all')
```

```python
from nltk.tag import pos_tag


def posTag(corpus):
    corpusNouns =[]
    corpusAdjectives =[]
    corpusVerbs =[]
    for tokenized_sent in corpus:
            nouns = [token for token, pos in pos_tag(tokenized_sent) if pos.startswith('N')]
            verbs = [token for token, pos in pos_tag(tokenized_sent) if pos.startswith('V')]
            adjectives = [token for token, pos in pos_tag(tokenized_sent) if pos.startswith('J')]
#             print (nouns)
            
#             print("...")
#             print (verbs)
            
#             print("...")
#             print (adjectives)
            
#             print("...")
            
            corpusNouns.append(nouns)
            corpusAdjectives.append(adjectives)
            corpusVerbs.append(verbs)

    return (corpusNouns,corpusAdjectives, corpusVerbs)
```

```python
posCorpusNouns,posCorpusAdjectives, posCorpusVerbs = posTag(posCorpus)
negCorpusNouns,negCorpusAdjectives, negCorpusVerbs = posTag(negCorpus)
# print("positive..")
# print(posCorpusNouns)
# print("..............................................")
# print("negative")
```

```python
posAllNouns =[]
posAllVerbs =[]
posAllAdjectives =[]
for posNouns in posCorpusNouns:
    posAllNouns.extend(posNouns)
for posVerb in posCorpusVerbs:
    posAllVerbs.extend(posVerb)
for posAdj in posCorpusAdjectives:
    posAllAdjectives.extend(posAdj)
    
posWords =[]
for posWord in posCorpus:
    posWords.extend(posWord)



negWords =[]
for negWord in negCorpus:
    negWords.extend(negWord)

```

```python
negAllNouns =[]
negAllVerbs =[]
negAllAdjectives =[]
for negNouns in negCorpusNouns:
    negAllNouns.extend(negNouns)
for negVerb in negCorpusVerbs:
    negAllVerbs.extend(negVerb)
for negAdj in negCorpusAdjectives:
    negAllAdjectives.extend(negAdj)
```

```python
onlyPosNouns = set(posAllNouns) - set(negAllNouns)
onlyPosAdjectives = set(posAllAdjectives) - set(negAllAdjectives)
onlyPosVerbs = set(posAllVerbs) - set(negAllVerbs)
```

```python
onlyNegNouns = set(negAllNouns) -set(posAllNouns) 
onlyNegAdjectives = set(negAllAdjectives) - set(posAllAdjectives)
onlyNegVerbs = set(negAllVerbs) - set(posAllVerbs) 
```
