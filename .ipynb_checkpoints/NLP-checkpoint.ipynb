```python

# library to clean data 
import re  
  
# Natural Language Tool Kit 
import nltk  
  
nltk.download('stopwords') 
  
# to remove stopword 
from nltk.corpus import stopwords 
  
import string
```

```python

corpus = []  
lemmatizer=nltk.stem.wordnet.WordNetLemmatizer()
# 1000 (reviews) rows to clean 

def makeCorpus(self, reviews)
    for review in reviews:  

        # column : "Review", row ith 
        #review = re.sub('[^a-zA-Z]', ' ', posReview)  

        # convert all cases to lower cases 
        review = review.lower()  

        # split to array(default delimiter is " ") 
        review = review.split()  



        # loop for stemming each word 
        # in string array at ith row     
        review = [lemmatizer.lemmatize(word) for word in review 
                    if not word in set(stopwords.words('english'))]  

        # rejoin all string array elements 
        # to create back into a string
        review1=[]
        for word in review:
            for ch in string.punctuation:                                                                                                     
                word = word.replace(ch, "")

            review1.append(word)
        review1 = [x for x in review1 if x]
        #print(review1)


        # append each string to create 
        # array of clean text  
        corpus.append(review1)  
    return corpus
```

```python
##downloading the nltk packages
import nltk
nltk.download('all')
```

```python
from nltk.tag import pos_tag

posCorpusNouns =[]
posCorpusAdjectives =[]
posCorpusVerbs =[]
for tokenized_sent in corpus:
        nouns = [token for token, pos in pos_tag(tokenized_sent) if pos.startswith('N')]
        verbs = [token for token, pos in pos_tag(tokenized_sent) if pos.startswith('V')]
        adjectives = [token for token, pos in pos_tag(tokenized_sent) if pos.startswith('J')]
        if nouns:
            posCorpusNouns.append(nouns)
        if adjectives:
            posCorpusAdjectives.append(adjectives)

        if verbs:
            posCorpusVerbs.append(verbs)

posCorpusNouns
```
